```python
import networkx as nx
g=nx.Graph()
g.add_edges_from([('A','B'),('A','H'),('H','G'),('I','G'),('I','F'),('G','F'),('F','D'),('D','C'),('B','C'),('A','G'),('C','E')])
```

```python
import matplotlib.pyplot as plt
degrees = [g.degree(n) for n in g.nodes()]
```

```python
print(degrees)
```

```python
plt.hist(degrees)
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
g=nx.DiGraph()
g.add_edges_from([('A','B'),('A','H'),('H','G'),('I','G'),('I','F'),('G','F'),('F','D'),('D','C'),('B','C'),('A','G'),('C','E')])
```

```python
degrees = [g.in_degree(n) for n in g.nodes()]
```

```python
plt.hist(degrees)
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
degrees = [g.out_degree(n) for n in g.nodes()]
```

```python
plt.hist(degrees)
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```
